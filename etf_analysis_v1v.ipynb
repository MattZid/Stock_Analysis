```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def etf_summary():
    try:
        print("Provide ETF tickers to review:")
        user_tick = input()
        user_tick = str.upper(user_tick).strip()
        print()
        etf = yf.Ticker(user_tick)
        etf_data = etf.funds_data
        etf_data.description
        print(f'Description of {user_tick}:')
        print(etf_data.description)
        print()
        print("Sector Weightings:")
        sector_data = etf_data.sector_weightings
        etf_df = pd.DataFrame.from_dict(sector_data,orient='index').reset_index()
        etf_df.rename(columns={'index':'Sector',0:'Percentage Holding'},inplace=True)
        etf_df = etf_df.sort_values(by='Percentage Holding',ascending=False,ignore_index=True).copy()
        print(etf_df)
        print()
        print("Top Holdings:")
        etf_df = etf_data.top_holdings
        etf_df.sort_values(by='Holding Percent',ascending=False,inplace=True)
        etf_df.reset_index()
        print(etf_df)
        top_equity = etf_df['Holding Percent'].aggregate('sum')
        print()
        print(f"The top stocks account for {top_equity} percent of this ETF's portfolio value.")
    except:
        print()
        print("Error.  ETF may not hold securities that can be summarized.")
        print("Please confirm input is a valid ETF ticker.")
```

```python
etf_summary()
```
